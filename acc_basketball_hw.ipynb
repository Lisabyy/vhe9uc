```python
import pandas as pd

acc_basketball_2324_df = pd.read_csv('acc_players-2324F.csv', sep=',', header = 1)

print("ACC Basketball 23-24 Players:")
display(acc_basketball_2324_df.head())
```

```python
total_points = acc_basketball_2324_df['PTS'].sum()
print(f"Total points scored by all players: {total_points}")

top_player_mp = acc_basketball_2324_df[['Player', 'MP']].sort_values(by='MP', ascending=False).head(1)
print(f"Top player who has played the most minutes: {top_player_mp}")

top_rebounds = acc_basketball_2324_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print(f"Top 5 players with total rebounds: {top_rebounds}")
```

```python
players_over_500_minutes_df = acc_basketball_2324_df[acc_basketball_2324_df['MP'] > 500]
print("Players who played more than 500 minutes:")
display(players_over_500_minutes_df)

top_player_assists = players_over_500_minutes_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(1)
print(f"Top player with the highest total assists: {top_player_assists}")

top_assist_leads = acc_basketball_2324_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print(f"Top 3 assist leads: {top_assist_leads}")

top_shot_blockers = acc_basketball_2324_df[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
print(f"Top 3 shot blockers: {top_shot_blockers}")
```

```python
school_total_points = acc_basketball_2324_df.groupby('School')[['PTS']].sum().reset_index()
print("Total points scored by each school:")
display(school_total_points)

school_total_assists = acc_basketball_2324_df.groupby('School')[['AST']].sum().reset_index()
print("Total assists scored for each school:")
display(school_total_assists)

top_total_points = acc_basketball_2324_df.groupby('School')[['PTS']].sum().sort_values(by='PTS', ascending=False).head(3)
print("Top 3 schools:")
display(top_total_points)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_total_points = acc_basketball_2324_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)
plt.figure(figsize=(10,6))
sns.barplot(x='PTS', y='Player', data=top_total_points)
plt.title('Top 5 Players By Total Points Scored')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='FG%', y='WS', data=acc_basketball_2324_df)
plt.title('Correlation between FG% and Win Shares (WS)')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.show()
```
